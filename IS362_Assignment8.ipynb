### Load the data in the data Folder in the file auto-mpg.data into a pandas DataFrame.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

files = "https://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data"
```

```python
column_name = ['MPG','Cylinders', 'Displacement', 'Horsepower', 'Weight', 'Acceleration', 'ModelYear', 'Origin', 'Model']
```

```python
df = pd.read_csv(files, delim_whitespace = True, names = column_name)
```

```python
df.head()
```

### This transforms Bad Data Values and Merge Original Dataframe with Country Name Dataframe by CountryID

```python
new = df.replace('?', np.nan)

country = pd.DataFrame({'Origin': [1, 2, 3], 'CountryName': ['America', 'Asia', 'Europe']})

transform = new.merge(country, left_on= 'Origin', right_on= 'Origin')

new_transform = transform.astype({'Horsepower': 'float'})
```

### This shows the Number of Cars with Certain Number of Cylinders

```python
import seaborn as seans

seans.set(style="darkgrid")
seans.barplot(x=transform['Cylinders'].value_counts().index, y=transform['Cylinders'].value_counts())
```

### This updates X- and Y- Axes to Avoid Overcrowding Values

```python
hp_incs = []

end = new_transform['Horsepower'].max()

start = new_transform['Horsepower'].min()

increment = np.math.ceil(end/25)

for x in range(increment + 1):
    x *= 25
    hp_incs.append(x)

end = new_transform['Weight'].max()

start = new_transform['Weight'].min()

increment = np.math.ceil(end/1500)

wt_incs = []

for x in range(increment + 1):
    x *= 1500
    wt_incs.append(x)
```

## Plot Scatterplot Graph
### Scatterplot Graph 1: The Basic Analysis Between Horsepower and Weight in X-Y Axis

```python
import matplotlib as plt
```

```python
fig_dims = (20, 4)
fig, ax = plt.pyplot.subplots(figsize = fig_dims)
ax.set(xlabel = hp_incs, ylabel = wt_incs)
seans.scatterplot(x = new_transform['Horsepower'], y=new_transform['Weight'], ax = ax)
```

### Scatterplot Graph 2: Adding Hue to Differentiate Country of Manufacturer

```python
fig_dims = (20, 4)
fig, ax = plt.pyplot.subplots(figsize = fig_dims)
ax.set(xlabel = hp_incs, ylabel = wt_incs)
seans.scatterplot(x = new_transform['Horsepower'], y = new_transform['Weight'], hue = new_transform['CountryName'], ax = ax)
```

### Scatterplot Graph 3: Adding Cylinders to Size of Plots and Analyzing MPG vs. Weight (Instead of Horsepower vs. Weight)

```python
fig_dims = (20, 10)
fig, ax = plt.pyplot.subplots(figsize = fig_dims)
ax.set(xlabel = hp_incs, ylabel = wt_incs)
seans.scatterplot(x = new_transform['MPG'], y = new_transform['Weight'], size = new_transform['Cylinders'], sizes = (20, 200), hue = new_transform['CountryName'], ax = ax)
```
